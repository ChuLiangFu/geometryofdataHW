Problem 2

For the metric space X, the topology can be formulated by the metric induced basis $B(x,r)$ for $ {\forall} x {\in} X $ and ${\forall} r {\in} \mathcal{R}$. So for$X \times X$, the topology is formulated by the basis:
\begin{align*}
B(x_1,r_1) \times B(x_2,r_2)
\end{align*}
So $A \times A$ should equip the corresponding subspace topology which has this kind of form basis based on the excercise 2 in section 1:
\begin{align*}
\{{B(x_1,r_1) \cap A}\} \times \{{B(x_2,r_2) \cap A \}}
\end{align*}
Where $ {\forall} x_1,x_2 {\in} X $ and ${\forall} r_1,r_2 {\in} \mathcal{R}$

For this excercise 2 in section 1 of note. Notice the definition of subspace topology is taking the intersection between the open sets from the original topology for the space X and subsapce Y, and notice the relation between topology and its basis is taking all kinds of union of basis to get the topology, it's easy to see the basis of subspace topology is nothing but the original space's basis of topology intersects with the subspace Y. This is because: $U \cap Y = (\cup_i B_i) \cap Y =\cup_i(B_i \cap Y) $.

On the other hand, $A \times A$ has another topology induced by the restricted metric on A, $d{\mid}_{A \times A}$. Then this topology is formulated by the basis:
\begin{align*}
B_A(x_3,r_3) \times B_A(x_4,r_4)
\end{align*}
Where $ {\forall} x_1,x_2 {\in} X $ and ${\forall} r_1,r_2 {\in} \mathcal{R}$ and $B_A$ represents the open ball equiped the metric restricted on A.

Then we would like to show these two basis are actually the same so that the induced topology should be the same as well.
we extend the $B_A(x_3,r_3) \times B_A(x_4,r_4)$ as $\{y \in A:d(x_3,y)<r_3 \} \times \{y \in A:d(x_4,y)<r_4 \}$. For $\{{B(x_1,r_1) \cap A}\} \times \{{B(x_2,r_2) \cap A \}}$, actually, it's $\{{\{y \in X:d(x_1,y)<r_1 \} \cap A}\} \times \{{\{y \in X:d(x_2,y)<r_2 \} \cap A \}}$. Take the intersection operator on the subspace A, then it becomes $\{{\{y \in A:d(x_1,y)<r_1 \} }\} \times \{{\{y \in A:d(x_2,y)<r_2 \}  \}}$, as the $x_1,x_2,x_3,x_4,r_1,r_2,r_3,r_4$ is selected in arbitrary, these two sets should be the same as each other which is exactly what we want.

Remark: this problem basically clarify the relation between the topology induced from subspace and the topology induced from the restricted metric on the subset. They should be the same as each other.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
#sample function to generate the random point on sphere
def samplefunc(n,d):
    t = d + 1
    x = np.zeros([n,t])
    y = np.zeros([n,t])
    for i in range(n):
        dist0 = 0
        for j in range(t):
            x[i,j] = np.random.normal(0,1)
            dist0 = dist0 + x[i,j]*x[i,j]
            
        dist00 = math.sqrt(dist0)
    
        for k in range(t):
            y[i,k]=x[i,k]/dist00    
    
    return y


dist = np.zeros([10000])

ax = samplefunc(10000,2)
# to calculate the arc length
for i in range(10000):
    dist1 = 0
    for j in range(3):
        if j!=0:
            dist1 = dist1 + ax[i,j]*ax[i,j]
    dist1 = math.sqrt(dist1)
    dist[i] = np.arccos(dist1)

plt.hist(dist)
plt.show()
```
