The idea is to subtract the row means and divide the norm within the matrix. Function [opa] is to use the svd to perform the Orthogonal Procrustes Analysis to get the aligned shape matrix with the target Fr´echet mean.  exponential map[exma] and log map[loma] is implemented at this special case. Because we perform the scaling, all of the shapes are actually sitting on a high dimensional sphere, so that we can simplify the calculation of exp map and log map by taking some simplified calculation. That's actually due to the sphere is what we are familiar manifold. The idea is to take the plane such that all the p, q and related vector are inside this plane with the great circle connected p,q. Then it is transfered into a 2D problem. For lop map calculation, the length is just the arccos of the inner product, the direction is got by taking the difference between q and its projection on p. For exp map, the v's length is $\theta$ and we can calculate q from the simple triangle relationship. For Fr´echet mean[freme], we take the gradient descent algorithm by using the log map and exp map within the iterations. For approximate principal geodesic analysis[pga], I am firstly using the [freme] to calculate Fr´echet mean and using log map to calculate all the required difference. Then I trasform all the data into the covariance matrix to do the svd to get the eigenanalysis.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def censcal(aaa):# removing its centroid and scaling 
    N =aaa.shape[0]
    ii=aaa.shape[1]   # take the dimension of the matrix
    jj=aaa.shape[2]   


    
    for k in range(N):
        meann=np.zeros(ii)
        tot=np.zeros(ii)
        sqfro=0
        for i in range(ii):
            for j in range(jj):
                tot[i] = tot[i] + aaa[k,i,j]        # sum up

        
        meann[i] = tot[i]/jj        #take the row mean

        for i in range(ii):
            for j in range(jj):
                aaa[k,i,j]=aaa[k,i,j]-meann[i] #Remove translation

        
        for i in range(ii):
            for j in range(jj):
                sqfro = sqfro + aaa[k,i,j]*aaa[k,i,j]  # sume the square up
        fronor = np.sqrt(sqfro)    #take the Frobenius norm
        
        for i in range(ii):
            for j in range(jj):
                aaa[k,i,j]=aaa[k,i,j]/fronor  #divide Frobenius norm
        
    return aaa


def opa(aa,bb): #Orthogonal Procrustes Analysis


    cc=np.dot(bb,np.transpose(aa))   #get BA^T
    u,sigma,vt = np.linalg.svd(cc)  #take svd
    R=np.dot(u,vt)                  #get R
    
    return R



def exma(aa,vv):#Compute the exponential map
    ii=vv.shape[0]   # take the dimension of the matrix
    jj=vv.shape[1]   



    sqfro = 0
    for i in range(ii):
        for j in range(jj):
            sqfro = sqfro + vv[i,j]*vv[i,j]  # sume the square up
    
    uu = vv
    theta = np.sqrt(sqfro)    #take the Frobenius norm    
    for i in range(ii):
        for j in range(jj):
            uu[i,j] = uu[i,j]/theta  # sum the square up
            

    costhe = np.cos(theta)
    sinthe = np.sin(theta)
    exm = aa*costhe+uu*sinthe
    ttt=np.linalg.norm(exm)
    print(exm)
    print(ttt)
    return exm



def loma(aa,bb):#Compute the log map, give tangent vector
    ii=aa.shape[0]   # take the dimension of the matrix
    jj=aa.shape[1]  
    inner = 0
    for i in range(ii):
        for j in range(jj):
            inner = inner + aa[i,j]*bb[i,j]
    
    #inner product
    
    length = np.arccos(inner)
    

    direction =bb-aa*inner
    
    lom = direction/np.linalg.norm(direction)*length
    
    return lom

def freme(aaa):   # Fr´echet mean
    
    N = aaa.shape[0]
    mu=aaa[1,:,:]
    dmu = 1
    admu = 1
    while admu > 0.0001:
        dmu = 0
        for m in range(N):
            RR = opa(aaa[m,:,:],mu)
            align = np.dot(RR,aaa[m,:,:])
            dmu = dmu + loma(mu,align)
        dmu = dmu / N
        admu = np.linalg.norm(dmu)
        print(admu)
        mu = exma(mu,dmu)
        
    return mu

    
def pga(aaa):
    N = aaa.shape[0]
    ii=aaa.shape[1]   # take the dimension of the matrix
    jj=aaa.shape[2]  
    u = np.zeros([N,ii,jj])
    kk = ii*jj
    v = np.zeros([N,kk])
    S = np.zeros([kk,kk])
    frem = freme(aaa) #get the mean
    for m in range(N):
        RR = opa(aaa[m,:,:],frem)
        align = np.dot(RR,aaa[m,:,:])
        u[m,:,:] = loma(frem,align)         #get the difference between mean and each point within tangent space
    # transfer the shape matrix into a vector
    for m in range(N):
        k = 0
        for i in range(ii):
            for j in range(jj):
                v[m,k] = u[m,i,j]
                k = k + 1
    # get the cov matrix
    for m in range(N):
        for sk1 in range(kk):
            for sk2 in range(kk):
                S[sk1,sk2]=S[sk1,sk2] + v[m,sk1]*v[m,sk2]/(N-1)
    uuu,sigma,vvvt = np.linalg.svd(S)
    
    return sigma
    
```

I consider this question is actually to ask the degree of freedom for the shape when we consider the equivalence coming from the translation, scaling, rotation and remove them to make up the quotient space.

1. for triangle in 2D here, there are three points and each point has two degree of freedom within 2D space. So the total degree of freedom is $2*3=6$. However, when we remove translation in 2D, we remove two degree of freedom, rotation for one degree of freedom, scaling for one degree of freedom, then it becomes: $2*3-2-1-1=2$. So the maximal number of modes here is 2.

2. for a set of objects with n points, similarly, we have 2*n=2n degree of freedom in total. Then remove translation, rotation and scaling, we have $2n-2-1-1=2n-4$. So the maximal number of modes here is $2n-4$.

Comment: actually, this reminds me the consider the equal partition theorem in statistical physics. Basically, it counts the degree of freedom for physical system and attach each degree of freedom some specific energy equally to describe the motion of physical system. You have to identify the actual total degree of freedom of the system first here as well.
